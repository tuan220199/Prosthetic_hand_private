```python
import matplotlib.pyplot as plt

# Data for each model
logistic_regression = [
    0.8914261858746128,
    0.8768556899711393,
    0.720115622875975,
    0.6266060361896731,
    0.5889465212719254 
]

# Multiply all values by 100
logistic_regression = [value * 100 for value in logistic_regression]

ffnn = [
    0.9338661481356798,
    0.9182078585699743,
    0.9090094868514337,
    0.8744497669579533,
    0.8595948210872656
]

# Multiply all values by 100
ffnn = [value * 100 for value in ffnn]

self_supervised = [
    0.9298203587532043,
    0.889467696348826,
    0.8885900179545084,
    0.7390552163124084,
    0.8155825138092041
]

# Multiply all values by 100
self_supervised = [value * 100 for value in self_supervised]

# Labels for the x-axis
labels = ['0 shift', '1 shift', '2 shifts', '3 shifts', '4 shifts']
x = range(len(labels))

# Plotting
plt.figure(figsize=(10, 6))

# Plotting the data points and trend lines
plt.plot(x, logistic_regression, marker='o', color='r', label='Logistic Regression')
plt.plot(x, ffnn, marker='o', color='g', label='FFNN')
plt.plot(x, self_supervised, marker='o', color='b', label='Self-supervised')

# Adding titles and labels
plt.title('Approach 3: Comparison between 3 Learning models for different shifts around cases')
plt.xlabel('Number of shifts around')
plt.ylabel('Accuracy %')

# Adding grid
plt.grid(True)
# Adding x-axis labels
plt.xticks(x, labels)

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
